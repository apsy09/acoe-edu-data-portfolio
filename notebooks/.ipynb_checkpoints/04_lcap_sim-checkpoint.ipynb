```python
from pathlib import Path
import pandas as pd
import yaml

ROOT = Path.cwd().resolve().parent
CONF = ROOT / "projects" / "04_lcap-strategic-plan-sim" / "config"
CONF.mkdir(parents=True, exist_ok=True)

goals_yaml = CONF / "goals.yaml"
interv_csv  = CONF / "interventions.csv"

# --- sensible defaults (edit later) ---
default_goals = {
    "district": "San Francisco County Office of Education",  # or "Burbank Unified"
    "year": "latest",         # or set an int, e.g., 2024
    "metric": "chronic_absent_rate",
    "target_mode": "relative", # "relative" or "absolute"
    "target_value": 0.15,      # 15% relative reduction (or an absolute fraction if mode=absolute)
    "focus_subgroups": ["English Learners","Students with Disabilities","Socioeconomically Disadvantaged"],
    "equity_gap_goal_pp": 0.02  # reduce subgroup gaps by 2.0 percentage points over a year
}

default_interventions = pd.DataFrame([
    {"lever":"Tier1_Family_Engagement","applies_to":"All","annual_effect_pp":1.0,"cost_estimate":"Low","notes":"Texting nudges; positive attendance framing"},
    {"lever":"Tier2_3_6_9_CheckIns","applies_to":"All","annual_effect_pp":0.8,"cost_estimate":"Medium","notes":"Structured outreach at 3/6/9 absences"},
    {"lever":"Transportation_Support","applies_to":"Socioeconomically Disadvantaged","annual_effect_pp":0.7,"cost_estimate":"Medium","notes":"Bus passes; routing support"},
    {"lever":"EL_Tutoring","applies_to":"English Learners","annual_effect_pp":0.5,"cost_estimate":"Medium","notes":"After-school targeted supports"},
    {"lever":"SPED_Case_Management","applies_to":"Students with Disabilities","annual_effect_pp":0.6,"cost_estimate":"High","notes":"Case manager cadence + family outreach"},
    {"lever":"Health_Mental_Referrals","applies_to":"All","annual_effect_pp":0.4,"cost_estimate":"Medium","notes":"Clinic/telehealth blocks; referrals"}
])

# Write files only if theyâ€™re missing
if not goals_yaml.exists():
    goals_yaml.write_text(yaml.safe_dump(default_goals, sort_keys=False), encoding="utf-8")
    print("Created", goals_yaml)

if not interv_csv.exists():
    default_interventions.to_csv(interv_csv, index=False)
    print("Created", interv_csv)

print("Config ready at:", CONF.as_posix())
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
import yaml
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

ROOT = Path.cwd().resolve().parent
PROCESSED = ROOT / "data" / "01_caldashboard" / "processed"
CONF = ROOT / "projects" / "04_lcap-strategic-plan-sim" / "config"
ASSETS = ROOT / "projects" / "04_lcap-strategic-plan-sim" / "assets"
ASSETS.mkdir(parents=True, exist_ok=True)

# --- data ---
wide_fp = PROCESSED / "kpi_chronic_wide_all_levels.csv"
assert wide_fp.exists(), f"Missing processed file: {wide_fp}"
wide = pd.read_csv(wide_fp)

# --- configs ---
goals_fp = CONF / "goals.yaml"
interv_fp = CONF / "interventions.csv"
assert goals_fp.exists(), f"Missing {goals_fp} (run the setup cell above)"
assert interv_fp.exists(), f"Missing {interv_fp} (run the setup cell above)"

with open(goals_fp, "r") as f:
    goals = yaml.safe_load(f)
interv = pd.read_csv(interv_fp)

# typing
for c in ("year","cohort","chronic_absent_rate","gap_vs_all"):
    if c in wide.columns:
        wide[c] = pd.to_numeric(wide[c], errors="coerce")

for c in ("annual_effect_pp",):
    if c in interv.columns:
        interv[c] = pd.to_numeric(interv[c], errors="coerce")
```

```python
DISTRICT = goals["district"]
YEAR = goals.get("year", "latest")
if YEAR == "latest":
    YEAR = int(wide.loc[wide["geo_level"].eq("district") & wide["district_name"].eq(DISTRICT), "year"].max())

df = wide[(wide["geo_level"].eq("district")) & (wide["district_name"].eq(DISTRICT)) & (wide["year"].eq(YEAR))].copy()
if df.empty:
    raise ValueError(f"No rows for {DISTRICT} in {YEAR}")

# baselines
all_row = df[df["subgroup"].astype(str).str.lower().isin(["all","all students"])].copy()
if all_row.empty:
    raise ValueError("Missing 'All Students' row needed for targets.")
baseline_all = float(all_row["chronic_absent_rate"].iloc[0])

sub = df[~df["subgroup"].astype(str).str.lower().isin(["all","all students"])].copy()
sub = sub[["subgroup","cohort","chronic_absent_rate","gap_vs_all"]].copy()
sub = sub.sort_values("chronic_absent_rate", ascending=False)

baseline_table = pd.concat([
    pd.DataFrame([{"subgroup":"All Students","cohort":float(all_row["cohort"].iloc[0]),"chronic_absent_rate":baseline_all,"gap_vs_all":0.0}]),
    sub
], ignore_index=True)

baseline_table
```

```python
mode = goals.get("target_mode","relative")
val = float(goals.get("target_value", 0.10))
if mode == "absolute":
    target_all = val  # fraction
else:
    target_all = baseline_all * (1 - val)  # e.g., 15% reduction

equity_pp = float(goals.get("equity_gap_goal_pp", 0.02))
focus = set(goals.get("focus_subgroups", []))

targets = baseline_table.copy()
targets["target_rate"] = np.where(
    targets["subgroup"].eq("All Students"),
    target_all,
    np.clip(target_all + np.maximum(targets["gap_vs_all"] - equity_pp, 0.0), 0, 1)
)
targets["delta_needed"] = (targets["chronic_absent_rate"] - targets["target_rate"]).round(4)
targets
```

```python
# interventions: treat All as applying to every subgroup + All Students row
iv = interv.copy()
iv["annual_effect_frac"] = iv["annual_effect_pp"] / 100.0

def applies(row_subgroup, applies_to):
    if applies_to == "All": return True
    return str(row_subgroup) == str(applies_to)

# build an impact table: each row gets sum of relevant levers
impact = []
for _, r in targets.iterrows():
    subtotal = iv[iv["applies_to"].apply(lambda a: applies(r["subgroup"], a))]["annual_effect_frac"].sum()
    impact.append(subtotal)

targets["lever_impact_est"] = np.round(impact, 4)

# scenarios: conservative (50%), expected (100%), stretch (130%) of lever impact
scenarios = {
    "conservative": 0.5,
    "expected": 1.0,
    "stretch": 1.3
}

out = {}
for name, mult in scenarios.items():
    t = targets.copy()
    t["projected_rate"] = np.clip(t["chronic_absent_rate"] - t["lever_impact_est"]*mult, 0, 1).round(4)
    t["shortfall"] = (t["projected_rate"] - t["target_rate"]).round(4)  # positive = still above target
    out[name] = t

out["expected"].head()
```

```python
import pandas as pd

for name, t in out.items():
    t.to_csv(ASSETS / f"lcap_projection_{name}_{YEAR}_{DISTRICT.lower().replace(' ','_')}.csv", index=False)

summary = pd.DataFrame({
    "scenario": list(out.keys()),
    "meets_all_target": [bool((d.loc[d["subgroup"].eq("All Students"),"shortfall"] <= 0).all()) for d in out.values()],
    "avg_shortfall_pp": [float((d["shortfall"].clip(lower=0).mean())*100) for d in out.values()]
})
summary.to_csv(ASSETS / f"lcap_projection_summary_{YEAR}_{DISTRICT.lower().replace(' ','_')}.csv", index=False)

summary
```

```python
def plot_all_waterfall(tdf, label):
    row = tdf[tdf["subgroup"].eq("All Students")].iloc[0]
    base = row["chronic_absent_rate"]; proj = row["projected_rate"]; tgt = row["target_rate"]

    plt.figure()
    plt.bar(["Baseline","Projected","Target"], [base, proj, tgt])
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1.0))
    plt.title(f"{DISTRICT} â€” All Students ({YEAR}) [{label}]")
    plt.tight_layout()
    out_png = ASSETS / f"waterfall_all_{label}_{YEAR}_{DISTRICT.lower().replace(' ','_')}.png"
    plt.savefig(out_png, dpi=150); plt.show()
    print("Saved:", out_png)

plot_all_waterfall(out["expected"], "expected")
```

```python
t = out["expected"].copy()
subg = t[~t["subgroup"].eq("All Students")].sort_values("projected_rate", ascending=False)

plt.figure()
plt.barh(subg["subgroup"], subg["projected_rate"])
plt.plot(subg["target_rate"], subg["subgroup"], "o")  # target markers
plt.gca().xaxis.set_major_formatter(PercentFormatter(1.0))
plt.xlabel("Rate"); plt.title(f"{DISTRICT} â€” Subgroup Projected vs Target ({YEAR})")
plt.tight_layout()
out_png = ASSETS / f"subgroups_proj_vs_target_{YEAR}_{DISTRICT.lower().replace(' ','_')}.png"
plt.savefig(out_png, dpi=150); plt.show()
print("Saved:", out_png)
```

```python
md = []
md.append(f"# LCAP Planning One-Pager â€” {DISTRICT} ({YEAR})\n")
md.append("## Metric\n- Chronic Absenteeism (fraction; lower is better)\n")
md.append("## Baselines\n")
row_all = targets[targets["subgroup"].eq("All Students")].iloc[0]
md.append(f"- All Students baseline: {row_all['chronic_absent_rate']:.3f}\n")
md.append("## Targets\n")
md.append(f"- All Students target: {row_all['target_rate']:.3f} "
          f"({goals.get('target_mode','relative')}={goals.get('target_value')})\n")
md.append(f"- Equity gap reduction goal: {goals.get('equity_gap_goal_pp')} pp\n")
md.append("## Interventions (expected combined annual impact)\n")
for _, r in interv.iterrows():
    md.append(f"- {r['lever']}: {r['annual_effect_pp']} pp (applies to: {r['applies_to']})\n")
md.append("\n## Projection â€” Expected Scenario\n")
row_all_exp = out["expected"][out["expected"]["subgroup"].eq("All Students")].iloc[0]
md.append(f"- Projected All Students rate: {row_all_exp['projected_rate']:.3f} "
          f"(shortfall vs target: {row_all_exp['shortfall']:.3f})\n")
md.append("\n## Subgroups needing attention (top 5 shortfalls)\n")
need = out["expected"].copy()
need = need[~need["subgroup"].eq("All Students")].sort_values("shortfall", ascending=False).head(5)
for _, r in need.iterrows():
    md.append(f"- {r['subgroup']}: shortfall {max(r['shortfall'],0):.3f}\n")
md.append("\n## Notes on Privacy & Governance\n- Uses aggregated public data; no PII. Small-N suppression applies (<10 cohort).\n")

text = "".join(md)
out_md = ASSETS / f"lcap_onepager_{YEAR}_{DISTRICT.lower().replace(' ','_')}.md"
out_md.write_text(text, encoding="utf-8")
print("Saved:", out_md)
```
